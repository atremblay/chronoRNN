```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma

def plot(std, chrono, legend, savefig=None, window=1, xlim=None, ylim=None):
    std_ = movingaverage(std, window)
    chrono_ = movingaverage(chrono, window)
    plt.plot(range(len(std_)), std_)
    plt.plot(range(len(chrono_)), chrono_)
    plt.legend(legend)
    plt.ylabel("NLL Loss")
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
        
    if savefig is not None:
        plt.savefig(savefig)
```

```python
with open('variable_copy_chrono_500/copyTask-LSTM-chrono-batch-30000-seed-0_0.json') as f:
    chrono_lstm = json.load(f)

with open('variable_copy_std_500/copyTask-LSTM-chrono-batch-29000-seed-0_0.json') as f:
    std_lstm = json.load(f)
plot(
    std_lstm['loss'], 
    chrono_lstm['loss'], 
    legend=['Constant 1', 'Chrono 500'],
    window=40, 
    ylim=(0,0.1), 
    savefig='variable_copy_500_repro.png'
)
```
